## Make a 10 Minute Driving Isochrone Around Pratt School of Information

```python
def make_isochrone(lat, lon, maxTime, timeUnit, distanceUnit, dateTime, travelMode):
    url = f"https://dev.virtualearth.net/REST/v1/Routes/Isochrones?waypoint={lat},{lon}&maxtime={maxTime}&timeUnit={timeUnit}&distanceUnit={distanceUnit}&dateTime={dateTime}&travelMode={travelMode}&key={BingMapsKey}"
    data = requests.get(url).json()
    points = []
    for i in data['resourceSets'][0]['resources'][0]['polygons'][0]['coordinates'][0]:
        point = shapely.geometry.Point(i[::-1])
        points.append(point)
    polygon = shapely.geometry.Polygon(points)
    polygon = gpd.GeoDataFrame(geometry=[polygon], crs=4326)
    return polygon
```

```python
lat = 40.738242
lon = -73.998971
maxTime = 10
timeUnit = 'minute'
distanceUnit = 'mi'
dateTime = '2023-03-01 17:30:00'
travelMode = 'driving'
```

```python
isochrone = make_isochrone(lat, lon, maxTime, timeUnit, distanceUnit, dateTime, travelMode)
```
